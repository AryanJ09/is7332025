```python
# prompt: Write a code to import the data from csv file present in drive

import pandas as pd

# Assuming your CSV file is in 'My Drive/data.csv'
file_path = '/content/drive/MyDrive/jobs.csv'  # Replace with your actual file path

df = pd.read_csv('/content/drive/MyDrive/jobs.csv')
```

```python
# Fill numeric columns with median
numeric_columns = df.select_dtypes(include=[np.number]).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())
```

```python
# Fill categorical columns with mode
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
#remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
yearly_trend = df.groupby(df['date'].dt.year).size()
plt.figure(figsize=(12, 6))
yearly_trend.plot(kind='bar')
plt.title('Job Postings by Year')
plt.xlabel('Year')
plt.ylabel('Number of Postings')
plt.show()
```

```python
# Monthly trend
monthly_trend = df.groupby(df['date'].dt.month).size()
plt.figure(figsize=(12, 6))
monthly_trend.plot(kind='bar')
plt.title('Job Postings by Month')
plt.xlabel('Month')
plt.ylabel('Number of Postings')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
# Day-of-week trend
day_of_week_trend = df.groupby(df['date'].dt.dayofweek).size()
plt.figure(figsize=(12, 6))
day_of_week_trend.plot(kind='bar')
plt.title('Job Postings by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Postings')
plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```
